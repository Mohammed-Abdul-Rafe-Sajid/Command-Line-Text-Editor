```python
class stack:
  def __init__(self):
    self.items = []

  def push(self, item):
    self.items.append(item)

  def pop(self):
    if len(self.items) <=0:
      return None
    else:
      return self.items.pop()
s=stack()
dl=stack()
main=[]
def delete():
    d=s.pop()
    dl.push(d)
    fobj=open("text.txt","w")
    l=[]
    for i in s.items:
      l.append(i)
    str1=''.join(l)
    fobj.write(str1)
    fobj.close()
while True:
  print("TEXT EDITOR")
  print("enter 1 for writing")
  print("enter 2 for displaying")
  print("enter 3 for deleting")
  print("enter 4 for undo")
  print("enter 5 to exit")
  n=int(input("Enter what task you want to do: "))
  main.append(n)

  if n==5:
    break

  if n==1: #writing

    fobj=open("text.txt","a")
    txt=input("Enter what you want to write: ")
    fobj.write(txt)

    for i in txt:
      s.push(i)
    fobj.close()


  if n==2: #displaying
    fobj=open("text.txt","r")
    txt=fobj.read()
    print("*********")
    print(txt)
    print("*********")

    print("")
    fobj.close()


  if n==3:#delete
    d=s.pop()
    dl.push(d)
    fobj=open("text.txt","w")
    l=[]
    #for i in s.items:
      #l.append(i)
    str1=''.join(s.items)
    fobj.write(str1)
    fobj.close()


  if n==4:
    for i in range(len(main)):
      if main[len(main)-i-1]==1:
        delete()
        break
      elif main[len(main)-i-1] ==3:
        fobj=open("text.txt","a")
        txt=dl.pop()
        if txt==None:
          break
        fobj.write(txt)
        fobj.close()
        break

```
